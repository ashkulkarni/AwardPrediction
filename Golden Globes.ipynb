## ML Models

```python
#Logistic Regression
def logisticReg(model_train, model_test):
    logisticClassify = linear_model.LogisticRegression()
    logisticClassify.fit(model_train.drop('next_golden_nominate', axis=1),model_train['next_golden_nominate'])
    logisticResults = logisticClassify.predict(model_test.drop('next_golden_nominate', axis=1))
    
    #evaluation
    #classification_evaluation(model_test['next_golden_nominate'],logisticResults)
    return f1_score(model_test['next_golden_nominate'],logisticResults,average=None)
    
```

```python
# Naive Bayes
def NaiveBayes(model_train, model_test):
    NBClassify = GaussianNB()
    NBClassify.fit(model_train.drop('next_golden_nominate', axis=1),model_train['next_golden_nominate'])
    NBResults = NBClassify.predict(model_test.drop('next_golden_nominate', axis=1))
    
    #evaluation
    #classification_evaluation(model_test['next_golden_nominate'],NBResults)
    return f1_score(model_test['next_golden_nominate'],NBResults,average=None)
```

```python
# Decision Tree
def DecisionTree(model_train, model_test):
    DTClassify = tree.DecisionTreeClassifier()
    DTClassify.fit(model_train.drop('next_golden_nominate', axis=1),model_train['next_golden_nominate'])
    DTResults = DTClassify.predict(model_test.drop('next_golden_nominate', axis=1))
    
    #evaluation
    #classification_evaluation(model_test['next_golden_nominate'],DTResults)
    return f1_score(model_test['next_golden_nominate'],DTResults,average=None)
```

```python
# K Nearest Neighbor
def KNN(model_train, model_test,n_neighbors):
    KNNClassify = neighbors.KNeighborsClassifier(n_neighbors)
    KNNClassify.fit(model_train.drop('next_golden_nominate', axis=1),model_train['next_golden_nominate'])
    KNNResults = KNNClassify.predict(model_test.drop('next_golden_nominate', axis=1))
    
    #evaluation
    #classification_evaluation(model_test['next_golden_nominate'],KNNResults)
    return f1_score(model_test['next_golden_nominate'],KNNResults,average=None)
```

### Resampling

```python
def resampleDataset(dataset):
    
    df_actual_winners = dataset.loc[(dataset['next_golden_nominate']==1),:].index.values.tolist()

    M2logisticClassify = linear_model.LogisticRegression()
    M2logisticClassify.fit(dataset.drop('next_golden_nominate', axis=1),dataset['next_golden_nominate'])
    M2logisticResults = M2logisticClassify.predict(dataset.drop('next_golden_nominate', axis=1))
    prob_dist = pd.DataFrame(M2logisticClassify.predict_proba(dataset.drop('next_golden_nominate', axis=1)))
    prob_dist.columns = ['non','win']
    high_prob = prob_dist.sort(columns='win', ascending=False).index.values.tolist()[:len(df_actual_winners)*2]
    
    new_index = list(set(df_actual_winners)|set(high_prob))
    return dataset.ix[new_index]
```

```python
def calculateWeightedMean(dataframe, f1_array):
    first_value = dataframe['next_golden_nominate'].iloc[0]
    second_value = first_value^1
    first_wt = len(dataframe.loc[(dataframe['next_golden_nominate']==first_value),:])
    second_wt = len(dataframe.loc[(dataframe['next_golden_nominate']==second_value),:])
    
    total_wt = first_wt + second_wt
    
    return 1.0*((f1_array[0]*first_wt)+(f1_array[1]*second_wt))/total_wt
```
